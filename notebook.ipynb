### After identifying the most relevant themes

```python
# Define the list of topics of interest
topics = ['Infrastructure: Communications',
         'Engaging the Private Sector (SABER)',
         'Tertiary Education (SABER)']

# Define the year range for exploring missing data: from 2000 to 2019 inclusive
years = np.arange(2000, 2020, 1)
years = [str(elem) for elem in years]
years = [elem for elem in years if elem in df3.columns]

for topic in topics:
    # Define a list of indicator names that belong to the topic
    topic_indicators = df5.loc[df5["Topic"] == topic]["Indicator Name"].unique()
    
    # Define a list to store the count of missing values for each indicator
    missing_counts = [0] * len(topic_indicators)
    
    # Calculate the number of missing values for each indicator in the topic
    for i in range(len(topic_indicators)):
        df = df3.loc[df3["Indicator Name"] == topic_indicators[i]][years]
        missing_counts[i] = df.isnull().sum().sum()
    
    # Display the percentage of missing data for each indicator in the current topic
    indicator_indices = [j for j in range(len(topic_indicators))]
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    missing_percentage = [count / (242 * len(years)) for count in missing_counts]
    ax.bar(indicator_indices, missing_percentage)
    plt.xlabel("Indicator index within the topic")
    plt.ylabel("Percentage of missing data")
    plt.title("Percentage of missing data for " + topic)
    plt.show()
```

#### Conclusion:

* According to the graphs above, the indicators from the three selected themes have a significant amount of missing values (100% except for one).

* I will therefore abandon this approach and move to the next one. The new idea involves calculating the percentage of missing data for all indicators, regardless of theme, and keeping only those indicators with a missing data percentage below a certain threshold.

### Exploring All Indicators

#### Selection of the Most Complete Indicators

```python
# List all indicators
indicators = df3["Indicator Name"].unique()

# Define the year range for exploring missing data: from 2010 to 2015 inclusive
years = [str(elem) for elem in np.arange(2010, 2016, 1) if str(elem) in df3.columns]

# Use groupby to compute missing percentages for all indicators at once
threshold = 0.5

# Group by indicator name and calculate missing percentage for each group
grouped = df3.groupby('Indicator Name')[years]

# Calculate missing percentage for each indicator
missing_percentage_series = grouped.apply(
    lambda g: g.isnull().sum().sum() / g.size
)

# Filter indicators with missing percentage below threshold
kept_indicators = missing_percentage_series[missing_percentage_series < threshold].index.tolist()
missing_perc = missing_percentage_series[missing_percentage_series < threshold].tolist()
kept_indicators[:10], len(kept_indicators)  # Show first 10 and total count
```

#### Selection of the Most Significant Indicators (by Keyword)

```python
# Define the list of keywords we are interested in
keywords = ["GDP", "internet", "enrolment"]

# Define the list that will contain the names of filtered indicators
filtered_indicators = []

# Filter indicators by keyword
for indicator in kept_indicators:
    test = False
    i = 0
    while test == False and i < len(keywords):
        if keywords[i].upper() in indicator.upper():
            test = True
            filtered_indicators.append(indicator)
        else:
            i += 1
filtered_indicators
```

# Choice of countries

```python
"""------------- Construction of the dataframe we will use to select countries -------------"""

# Definition of already selected years
years = [str(elem) for elem in np.arange(2010, 2016, 1) if str(elem) in df3.columns]

# Definition of the list of selected indicators
final_indicators = ['Internet users (per 100 people)',
                    'Enrolment in secondary education, both sexes (number)',
                    'Enrolment in tertiary education, all programmes, both sexes (number)',
                    'GDP per capita (constant 2005 US$)']

# Definition of regions that will not be part of our study
regions = ['Arab World', 'East Asia & Pacific', 'East Asia & Pacific (excluding high income)', 'Euro area',
           'Europe & Central Asia', 'Europe & Central Asia (excluding high income)', 'European Union',
           'Heavily indebted poor countries (HIPC)', 'High income', 'Latin America & Caribbean',
           'Latin America & Caribbean (excluding high income)', 'Least developed countries: UN classification',
           'Low & middle income', 'Low income', 'Lower middle income', 'Middle East & North Africa',
           'Middle East & North Africa (excluding high income)', 'Middle income', 'North America', 'OECD members',
           'South Asia', 'Sub-Saharan Africa', 'Sub-Saharan Africa (excluding high income)', 'Upper middle income', 'World']

# Definition of columns we are interested in
cols = ["Country Name", "Indicator Code", "Indicator Name"] + years

# Selection of rows (excluding regions and unnecessary indicators) and columns we are interested in
df = df3.loc[df3["Indicator Name"].isin(final_indicators)][cols]
df = df.loc[~df["Country Name"].isin(regions)]
# Add a column containing the average of each indicator for each country over the studied year range
df["Average Over Years"] = df.mean(axis=1, numeric_only=True)

df.head()
```

```python
"""--------------- Select the list of most interesting countries for each indicator ----------------"""

# Define a dictionary to store the list of most interesting countries for each indicator
best_countries_per_indicator = {}

# Select the list of most interesting countries for each indicator
for indicator in final_indicators:
    # Select rows corresponding to the current indicator, each row corresponds to a country
    df_indicator = df.loc[df["Indicator Name"] == indicator]
    
    # Select the 25 most interesting countries for each indicator
    df_indicator = df_indicator.nlargest(n=25, columns="Average Over Years")
    
    # Add the list of most interesting countries for the current indicator to the dictionary
    best_countries_per_indicator[indicator] = df_indicator["Country Name"].unique()
    
    # Display results as a graph
    df_indicator.plot(x="Country Name", y="Average Over Years", kind="bar", figsize=(15, 8), title=indicator)
```

```python
# Concatenation of the most interesting countries for each indicator (with redundancy)
all_best_countries = []
for elem in list(best_countries_per_indicator.values()):
    all_best_countries.append(list(elem))
all_best_countries = [elem for sublist in all_best_countries for elem in sublist]

# Creation of a dictionary in the following form: {country: number_of_occurrences_in_all_best_countries}
all_best_countries_occ = {}
for country in set(all_best_countries):
    all_best_countries_occ[country] = all_best_countries.count(country)

# Selection of countries that have an occurrence count greater than or equal to 3
selected_best_countries_occ = []
for item in all_best_countries_occ:
    if all_best_countries_occ[item] >= 3:
        selected_best_countries_occ.append(item)
selected_best_countries_occ
```

```python
"""---------------- Construction of the final dataframe --------------------"""
df_finale = df.loc[df["Country Name"].isin(selected_best_countries_occ)]

# Visualization of the evolution of each indicator during the chosen period for each country
for indicator in df_finale["Indicator Code"].unique():
    plt.figure(figsize=(10, 4))
    plot_legend = []
    for country in selected_best_countries_occ:
        plot_legend.append(country)
        plt.plot(df_finale.loc[(df_finale["Country Name"] == country) & (df_finale["Indicator Code"] == indicator)][years].T)
        
    plt.xlabel("Years")
    plt.ylabel("Indicator value")
    plt.title(indicator)
    plt.legend(plot_legend)
    plt.show()
```
