### Après identification des thèmes les plus cohérents

```python
# définir la liste des topics d'intérêt
topics = ['Infrastructure: Communications', \
         'Engaging the Private Sector (SABER)', \
         'Tertiary Education (SABER)']

# définir l'intervalle des années sur lequel nous allons explorer les données manquantes: de 2010 à 2015 inclus
annees = np.arange(2000, 2020, 1)
annees = [str(elem) for elem in annees]
annees = [elem for elem in annees if elem in df3.columns]

for topic in topics:
    # définir une liste des noms des indicateurs qui constituent le topic
    topic_indicators = df5.loc[df5["Topic"]==topic]["Indicator Name"].unique()
    
    # définir une liste qui va contenir le nombre de datas manquantes pour chaque indicateur
    missing_counts = [0]*len(topic_indicators)
    
    # calculer le nombre de datas manquantes pour chaque indicateur du topic
    for i in range(len(topic_indicators)):
        df = df3.loc[df3["Indicator Name"]==topic_indicators[i]][annees]
        missing_counts[i] = df.isnull().sum().sum()
    
    # Affichage du taux de datas manquantes de chaque indicateur du topic en question
    ind_indicateurs = [j for j in range(len(topic_indicators))]
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    missing_percentage = [count/(242*len(annees)) for count in missing_counts]
    ax.bar(ind_indicateurs,missing_percentage)
    plt.xlabel("indice de l'indicateur dans le topic")
    plt.ylabel("pourcentage des données manquantes")
    plt.title("Pourcentage des données manquantes pour " + topic)
    plt.show()
```

#### Conclusion:

* D'après les graphes ci dessus, les indicateurs des 3 thèmes choisis représentent beaucoup de valeurs manquantes (100% à l'exception d'un seul)
* Je vais alors abondonner cette idée pour passer à une suivante. La nouvelle idée consiste à calculer le pourcentage des données manquantes pour tous les indicateurs, sans tenir compte du thème, pour ne garder que les indicateurs qui ont un pourcentage de missing data inférieur à un certain seuil.

### Exploration de tous les indicateurs

#### Sélection des indicateurs les plus remplis

```python
# lister tous les indicateurs
indicators = df3["Indicator Name"].unique()

# définir l'intervalle des années sur lequel nous allons explorer les données manquantes: de 2010 à 2015 inclus
annees = [str(elem) for elem in np.arange(2010, 2016, 1) if str(elem) in df3.columns]

# éliminer les indicateurs dont le taux de données manquantes est inférieurs au seuil choisi
held_indicators = []
missing_perc = []
seuil = 0.5

for indicator in indicators:
    df = df3.loc[df3["Indicator Name"]==indicator][annees]
    pourcentage = df.isnull().sum().sum()/df.size
    if (pourcentage<seuil):
        held_indicators.append(indicator)
        missing_perc.append(pourcentage)
```

#### Sélection des indicateurs les plus significatifs (par mot clé)

```python
# définition de la liste des mots clés qui nous intéresse
filtre = ["GDP", "internet", "enrolment"]

# définir la liste qui va contenir le nom des indicateurs filtrés
filtered_indicators = []

# filtrer les indicateurs par mot clé
for indicator in held_indicators :
    test = False
    i = 0
    while test==False and i<len(filtre):
        if filtre[i].upper() in indicator.upper():
            test = True
            filtered_indicators.append(indicator)
        else:
            i+=1
filtered_indicators
```

## Conclusion
* La liste filtered_indicators récapitule les indicateurs choisis selon 2 critères : le nombre des données manquantes et les mots clés.
* Je me suis basé sur cette liste pour construire la liste finale des indicateurs choisis qui sont : 'Internet users (per 100 people)', 'Enrolment in secondary education, both sexes (number)', 'Enrolment in tertiary education, all programmes, both sexes (number)', 'GDP per capita (constant 2005 US$)'

# Choix des pays

```python
"""------------- Construction de la dataframe que nous allons utiliser pour choisir les pays -------------"""

# Définition des années déjà sélectionnées
annees = [str(elem) for elem in np.arange(2010, 2016, 1) if str(elem) in df3.columns]

# Définition de la liste des indicateurs choisis
final_indicators = ['Internet users (per 100 people)', \
                    'Enrolment in secondary education, both sexes (number)', \
                    'Enrolment in tertiary education, all programmes, both sexes (number)', \
                    'GDP per capita (constant 2005 US$)']

# Définition des régions qui ne feront pas partie de notre étude
regions = ['Arab World', 'East Asia & Pacific', 'East Asia & Pacific (excluding high income)', 'Euro area', \
           'Europe & Central Asia', 'Europe & Central Asia (excluding high income)', 'European Union', \
           'Heavily indebted poor countries (HIPC)', 'High income', 'Latin America & Caribbean', \
           'Latin America & Caribbean (excluding high income)', 'Least developed countries: UN classification', \
           'Low & middle income', 'Low income', 'Lower middle income', 'Middle East & North Africa', \
           'Middle East & North Africa (excluding high income)', 'Middle income', 'North America', 'OECD members', \
           'South Asia', 'Sub-Saharan Africa', 'Sub-Saharan Africa (excluding high income)', 'Upper middle income', 'World']

# Définition des colonnes qui nous intéressent
cols = ["Country Name", "Indicator Code", "Indicator Name"] + annees

# Sélection des lignes (écarter les régions et les indicateurs non nécessaires) et des colonnes qui nous intéressent
df = df3.loc[df3["Indicator Name"].isin(final_indicators)][cols]
df = df.loc[~df["Country Name"].isin(regions)]
# Ajouter une colonne qui contient la moyenne de chaque indicateur pour chaque pays sur l'intervalle des années étudées
df["Average Over Years"] = df.mean(axis=1, numeric_only=True)

df.head()
```

```python
"""--------------- Choisir la liste des pays les plus intéressants pour chaque indic ----------------"""

# Définir un dictionnaire qui va contenir la liste des pays les plus intéressants pour chaque indicateur
best_countries_per_indicator = {}

# Choisir la liste des pays les plus intéressants pour chaque pays
for indicator in final_indicators:
    # Sélection des lignes qui correspondent à l'indicateur en question, chaque ligne correspond à un pays
    df_indicator = df.loc[df["Indicator Name"]==indicator]
    
    # Sélection des 25 pays les plus intéressants pour chaque indicateur
    df_indicator = df_indicator.nlargest(n=25, columns="Average Over Years")
    
    # Ajouter la liste des pays les plus intéressants pour l'indicateur en question au dictionnaire
    best_countries_per_indicator[indicator]=df_indicator["Country Name"].unique()
    
    # Afficher les résultats sous la forme d'un graphe
    df_indicator.plot(x="Country Name", y="Average Over Years", kind="bar", figsize=(15, 8), title=indicator)
```

```python
# concaténation des pays les plus intéressants pour chaque indicateur (avec redondance)
all_best_countries = []
for elem in list(best_countries_per_indicator.values()):
    all_best_countries.append(list(elem))
all_best_countries = [elem for sublist in all_best_countries for elem in sublist]

# création d'un dictionnaire de la forme suivante: {pays: nbre_d'occurrence_dans_all_best_countries}
all_best_countries_occ = {}
for country in set(all_best_countries):
    all_best_countries_occ[country] = all_best_countries.count(country)

# sélection des pays qui ont une occurrence supérieure à 3
selected_best_countries_occ = []
for item in all_best_countries_occ:
    if all_best_countries_occ[item]>=3:
        selected_best_countries_occ.append(item)
selected_best_countries_occ
```

```python
"""---------------- Construction de la dataframe finale --------------------"""
df_finale = df.loc[df["Country Name"].isin(selected_best_countries_occ)]

# visualisation de l'évolution de chaque indicateur durant la période choisie pour chaque pays
for indicator in df_finale["Indicator Code"].unique():
    plt.figure(figsize=(10, 4))
    plot_legend = []
    for country in selected_best_countries_occ:
        plot_legend.append(country)
        plt.plot(df_finale.loc[(df_finale["Country Name"]==country) & (df_finale["Indicator Code"]==indicator)][annees].T) 
        
    plt.xlabel("années")
    plt.ylabel("valeur de l'indicateur")
    plt.title(indicator)
    plt.legend(plot_legend)
    plt.show()
```
